```python
vaccine_df=pd.read_csv("covid_vaccine_statewise.csv")
```

```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True , axis=1)
```

```python
covid_df.head(10)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format='%m/%d/%Y')
```

```python
covid_df.head()
```

```python
#Findding Active Cases
covid_df["Active_Cases"]= covid_df["Confirmed"] - (covid_df["Cured"] + covid_df["Deaths"])
covid_df.tail()
```

```python
statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured"], index="State/UnionTerritory", aggfunc = max)
```

```python
statewise["Recovery Rate"]= statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"]= statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed" , ascending=False )
```

```python
statewise.style.background_gradient(cmap="Greens")
```

```python
#Top ten active cases states
top_ten=covid_df.groupby(by='State/UnionTerritory').max()[['Cured','Date']].sort_values(by=['Cured'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title("Top ten states with most Cured in India", size=27)
```

```python
#Top 10 states with MOST CURED

#GROUPING AND SORTTING DATA
top_ten=covid_df.groupby(by='State/UnionTerritory').max()[['Cured','Date']].sort_values(by=['Cured'],ascending=False).reset_index()

#FIGURE THE CHART SIZE
fig=plt.figure(figsize=(16,9))

#WIRTE THE CHART TITLE
plt.title("Top ten states with most Cured in India", size=27)

#DECLEAR THE CHART PARAMETERS
ax=sns.barplot(data= top_ten.iloc[0:10], y="Cured" , x="State/UnionTerritory",linewidth=2 , edgecolor='red')

plt.xlabel('state')
plt.ylabel('cured')
plt.show()
```

```python
#Top 10 states with MOST DEATHS
top_tend=covid_df.groupby(by='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top ten states with most Deaths in India", size=27)
ax=sns.barplot(data= top_tend.iloc[0:10], y="Deaths" , x="State/UnionTerritory",linewidth=2 , edgecolor='black')
plt.xlabel('state')
plt.ylabel('Deaths')
plt.show()
```

```python
#Top 10 states with MOST ACTIVE CASES
top_tenc=covid_df.groupby(by='State/UnionTerritory').max()[['Active_Cases','Date']].sort_values(by = ['Active_Cases'],ascending = False).reset_index()
fig=plt.figure(figsize=(30,9))
plt.title('Comparison Most Active_Cases in India on a larger scale',size=27)
ax=sns.barplot(data=top_tenc.iloc[:20],y='Active_Cases', x="State/UnionTerritory",linewidth=2 , edgecolor='red')
plt.xlabel('state')
plt.ylabel('Active_Cases')
plt.show()
```

```python
#Growth Trend in top 10 States
fig=plt.figure(figsize=(12,6))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Rajasthan','Chhattisgarh','West Bengal','Delhi','Andhra Pradesh','Uttar Pradesh','Tamil Nadu','Kerala','Karnataka','Maharashtra'])]
,
x='Date', y='Active_Cases', hue='State/UnionTerritory')

ax.set_title('Comparing Active Cases of Top 10 states in India', size=16)
```

```python
vaccine.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
vaccine
```

```python
Maxi_vaccine=vaccine.groupby('State')['Total'].sum().to_frame('Total')
Maxi_vaccine.sort_values('Total', ascending = False )[:10]
```
